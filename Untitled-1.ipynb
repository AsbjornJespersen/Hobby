```python
import yfinance as yf
import pandas as pd

# Your current holdings
portfolio = {
    'AAPL': 11,
    'MSFT': 5,
    'NVDA': 17,
    'CSCO': 23,
    'AMZN': 7,
    'GOOGL': 8,
    'TMUS': 5,
    'AVGO': 6,
    'META': 2,
    'NFLX': 1,
    'BRK-B': 2,
    'COST': 1,
}
nasdaq_weights = {
    'AAPL': 8.5,
    'MSFT': 7.6,
    'NVDA': 6.2,
    'AMZN': 5.5,
    'META': 4.2,
    'GOOGL': 3.5,
    'AVGO': 3.2,
    'COST': 2.4,
    'TMUS': 2.0,
    'CSCO': 1.9,
    'NFLX': 1.5,
    'BRK-B': 1.2,
    # You can add more tickers if needed
}

tickers = list(portfolio.keys())

# Download adjusted close prices
data = yf.download(tickers, period='1d', auto_adjust=True)

# If multiple tickers, 'Adj Close' becomes a column level
if isinstance(data.columns, pd.MultiIndex):
    prices = data['Close'].iloc[-1]
else:
    prices = data.iloc[-1]

# Create portfolio dataframe
df = pd.DataFrame({
    'Shares': pd.Series(portfolio),
    'Price': prices
})
df['Value'] = df['Shares'] * df['Price']
df['Weight (%)'] = 100 * df['Value'] / df['Value'].sum()

df = df.sort_values('Weight (%)', ascending=False)

display(df)
```

```python
# Step 1: Total USD value
total_usd = df['Value'].sum()

# Step 2: Get USD to DKK exchange rate using yfinance
fx_rate = float(yf.download('USDDKK=X', period='1d', interval='1m')['Close'].dropna().iloc[-1])

# Step 3: Convert to DKK
total_dkk = total_usd * fx_rate

print(f"Total Portfolio Value:")
print(f"USD: ${total_usd:,.2f}")
print(f"DKK: {total_dkk:,.2f} DKK (Exchange rate: {fx_rate:.2f})")
```

```python
# Add value in DKK per stock
df['Value (DKK)'] = df['Value'] * fx_rate

# Reorder for readability (optional)
df = df[['Shares', 'Price', 'Value', 'Value (DKK)', 'Weight (%)']]

# Sort by value (optional)
df = df.sort_values('Value (DKK)', ascending=False)

# Show the updated DataFrame
display(df)
```

```python
# Create DataFrame from Nasdaq-100 weights
nasdaq_df = pd.DataFrame.from_dict(nasdaq_weights, orient='index', columns=['Nasdaq100 Weight (%)'])
nasdaq_df.index.name = 'Ticker'

# Merge with your portfolio data
comparison = df.merge(nasdaq_df, left_index=True, right_index=True, how='outer').fillna(0)
comparison['Difference (%)'] = comparison['Weight (%)'] - comparison['Nasdaq100 Weight (%)']
comparison = comparison.sort_values('Difference (%)')

display(comparison)
```

```python
import matplotlib.pyplot as plt

# Pie chart of current portfolio
plt.figure(figsize=(8, 8))
plt.pie(df['Weight (%)'], labels=df.index, autopct='%1.1f%%', startangle=140)
plt.title('Your Portfolio Allocation')
plt.axis('equal')
plt.show()
```

```python
# Only show overlapping tickers
overlap = comparison[comparison['Nasdaq100 Weight (%)'] > 0]

# Bar chart
plt.figure(figsize=(12, 6))
x = overlap.index
plt.bar(x, overlap['Weight (%)'], label='Your Portfolio')
plt.bar(x, overlap['Nasdaq100 Weight (%)'], alpha=0.7, label='Nasdaq-100')
plt.xticks(rotation=45)
plt.ylabel('Weight (%)')
plt.title('Portfolio vs Nasdaq-100')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import requests

# ---- 1. Define your portfolio ----
portfolio = {
    'AAPL': 11,
    'MSFT': 5,
    'NVDA': 17,
    'CSCO': 23,
    'AMZN': 7,
    'GOOGL': 8,
    'TMUS': 5,
    'AVGO': 6,
    'META': 2,
    'NFLX': 1,
    'BRK-B': 2,
    'COST': 1,
}

# ---- 2. Fetch current prices ----
tickers = list(portfolio.keys())
data = yf.download(tickers, period='1d', auto_adjust=True)

# Handle multi-index for multiple tickers
if isinstance(data.columns, pd.MultiIndex):
    prices = data['Close'].iloc[-1]
else:
    prices = data.iloc[-1]

# ---- 3. Calculate portfolio weights ----
df = pd.DataFrame({
    'Shares': pd.Series(portfolio),
    'Price': prices
})
df['Value'] = df['Shares'] * df['Price']
df['Weight (%)'] = 100 * df['Value'] / df['Value'].sum()

# URL for Slickcharts Nasdaq-100
url = "https://www.slickcharts.com/nasdaq100"

# Use headers to mimic a browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.93 Safari/537.36"
}

# Make the request
response = requests.get(url, headers=headers)

# Parse HTML tables
tables = pd.read_html(response.text)

# Use the first table
nasdaq_df = tables[0]


# Clean up
nasdaq_df = nasdaq_df[['Symbol', 'Company', 'Portfolio%']]
nasdaq_df.columns = ['Ticker', 'Company', 'Nasdaq100 Weight (%)']
nasdaq_df.set_index('Ticker', inplace=True)
nasdaq_df['Nasdaq100 Weight (%)'] = (
    nasdaq_df['Nasdaq100 Weight (%)']
    .astype(str)             # in case it's not a string yet
    .str.replace('%', '')    # remove any percent signs
    .str.replace(',', '')    # remove any commas
    .astype(float)           # convert to float
)

# ---- 5. Merge your portfolio with Nasdaq-100 weights ----
comparison = df.merge(nasdaq_df[['Nasdaq100 Weight (%)']], left_index=True, right_index=True, how='outer').fillna(0)
comparison['Difference (%)'] = comparison['Weight (%)'] - comparison['Nasdaq100 Weight (%)']
comparison = comparison.sort_values('Nasdaq100 Weight (%)', ascending=False)

# ---- 6. Plot bar chart: Nasdaq-100 weights ----
plt.figure(figsize=(18, 6))
plt.bar(comparison.index, comparison['Nasdaq100 Weight (%)'], color='gray', label='Nasdaq-100')
plt.bar(comparison.index, comparison['Weight (%)'], alpha=0.7, label='Your Portfolio')
plt.xticks(rotation=90)
plt.ylabel('Weight (%)')
plt.title('Your Portfolio vs. Nasdaq-100 Allocation')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 1. Merge again to ensure latest weights
comparison = df.merge(nasdaq_df[['Nasdaq100 Weight (%)']], left_index=True, right_index=True, how='outer').fillna(0)
comparison['Difference (%)'] = comparison['Nasdaq100 Weight (%)'] - comparison['Weight (%)']

# 2. Total portfolio value
total_usd = df['Value'].sum()

# 3. Simulate adding $1000 to each stock and recompute the new weight
suggestions = []

for ticker, row in comparison.iterrows():
    current_value = row['Value']
    new_value = current_value + 1000
    new_total = total_usd + 1000
    new_weight = 100 * new_value / new_total
    new_diff = abs(new_weight - row['Nasdaq100 Weight (%)'])
    suggestions.append((ticker, new_diff))

# 4. Sort by best improvement (smallest remaining gap)
suggestions.sort(key=lambda x: x[1])

# 5. Display top suggestion
best_ticker, projected_gap = suggestions[0]
print(f" Suggested stock to buy for $1000: {best_ticker}")
print(f" Projected deviation after buy: {projected_gap:.2f}%")
```
